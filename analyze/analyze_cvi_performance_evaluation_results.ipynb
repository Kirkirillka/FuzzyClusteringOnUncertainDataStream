```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
cvi_df = pd.read_csv("cvi.csv")
performance_df = pd.read_csv("performance.csv")
```

```python
cvi_df["Name"] = cvi_df["Algorithm"] + "_" + cvi_df["Stream_Type"] + "_" + cvi_df["History_Mode"].apply(str)
```

```python
def bar_plot(df):
    metric_name = df["Metric"].iloc[0]
    plt.title(metric_name)
    sns.catplot(x="Algorithm", y="Value",hue="Stream_Type", data = df, kind="bar",
                height=4, aspect=.7)
```

```python
cvi_df.groupby("Metric", group_keys=True).apply(bar_plot)
```
